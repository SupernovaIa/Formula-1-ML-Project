## Track map

```python
session_map = session_qualy

# Telemetry is needed
session_map.load()
```

```python
lap = session_map.laps.pick_fastest()
pos = lap.get_pos_data()

circuit_info = session_map.get_circuit_info()

# Obtener el array con los puntos x e y
track = pos.loc[:, ('X', 'Y')].to_numpy()

# Convertir el ngulo de rotacin de grados a radianes
track_angle = circuit_info.rotation / 180 * np.pi

# Funcin para rotar puntos
def rotate(points, angle):
    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                                 [np.sin(angle),  np.cos(angle)]])
    return np.dot(points, rotation_matrix.T)

# Rotar el track
rotated_track = rotate(track, angle=track_angle)

# Crear la figura inicial
fig = go.Figure()

# Agregar la pista rotada
fig.add_trace(go.Scatter(x=rotated_track[:, 0], y=rotated_track[:, 1],
                         mode='lines',
                         line=dict(color='blue'),
                         name='Track'))

# Vector de desplazamiento
offset_vector = np.array([500, 0])

# Iterar sobre las esquinas
for _, corner in circuit_info.corners.iterrows():
    # Crear el texto de la esquina
    txt = f"{corner['Number']}{corner['Letter']}"

    # Convertir el ngulo de desplazamiento a radianes
    offset_angle = corner['Angle'] / 180 * np.pi

    # Rotar el vector de desplazamiento
    offset_x, offset_y = rotate(offset_vector, angle=offset_angle)

    # Aadir el desplazamiento a la posicin de la esquina
    text_x = corner['X'] + offset_x
    text_y = corner['Y'] + offset_y

    # Rotar la posicin del texto y el centro de la esquina
    text_x, text_y = rotate(np.array([text_x, text_y]), angle=track_angle)
    track_x, track_y = rotate(np.array([corner['X'], corner['Y']]), angle=track_angle)

    # Dibujar un crculo al lado de la pista
    fig.add_trace(go.Scatter(x=[text_x], y=[text_y],
                             mode='markers',
                             marker=dict(color='rgba(0, 128, 255, 0.6)', size=14),
                             name=f'Corner {txt}'))

    # Dibujar una lnea desde la pista hasta el crculo
    fig.add_trace(go.Scatter(x=[track_x, text_x], y=[track_y, text_y],
                             mode='lines',
                             line=dict(color='rgba(255, 0, 0, 0.5)', dash='dot'),
                             showlegend=False))

    # Aadir el texto de la esquina
    fig.add_trace(go.Scatter(x=[text_x], y=[text_y],
                             mode='text',
                             text=[txt],
                             textfont=dict(size=10, color='white'),
                             showlegend=False))

# Configurar el ttulo y el diseo de la figura
fig.update_layout(
    xaxis=dict(showticklabels=False, showgrid=False, zeroline=False),
    yaxis=dict(showticklabels=False, showgrid=False, zeroline=False),
    title=session.event['Location'],
    showlegend=False,
    template='plotly_dark',
)

# Hide grid lines
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)

fig.update_yaxes(scaleanchor="x", scaleratio=1)

# Mostrar la figura
fig.show()
```

### Qualy results

```python
drivers = session_qualy.laps['Driver'].unique()
```

```python
list_fastest_laps = []

for drv in drivers:
    drvs_fastest_lap = session.laps.pick_drivers([drv]).pick_fastest()
    list_fastest_laps.append(drvs_fastest_lap)
```

```python
from fastf1.core import Laps

fastest_laps = Laps(list_fastest_laps).sort_values(by='LapTime').reset_index(drop=True)
```

```python
pole_lap = fastest_laps.pick_fastest()
fastest_laps['LapTimeDelta'] = fastest_laps['LapTime'] - pole_lap['LapTime']
```

```python
team_colors = []

for _, lap in fastest_laps.iterlaps():
    color = fastf1.plotting.get_team_color(lap['Team'], session=session_qualy)
    team_colors.append(color)
```

```python
fig, ax = plt.subplots()
ax.barh(fastest_laps.index, fastest_laps['LapTimeDelta'],
        color=team_colors)
ax.set_yticks(fastest_laps.index)
ax.set_yticklabels(fastest_laps['Driver'])

# show fastest at the top
ax.invert_yaxis()

# draw vertical lines behind the bars
ax.set_axisbelow(True)
ax.xaxis.grid(True, which='major', linestyle='--', color='black', zorder=-1000)

lap_time_string = str(pole_lap['LapTime']).replace('0 days 00:', '').removesuffix('000')

plt.suptitle(f"{session_qualy.event['EventName']} {session_qualy.event.year} Qualifying\n"
             f"Fastest Lap: {lap_time_string} ({pole_lap['Driver']})")

plt.show()
```
